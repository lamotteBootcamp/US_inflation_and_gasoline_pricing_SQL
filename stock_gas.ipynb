```python
import pandas as pd
import numpy as np
```

```python
csv_path = "US_gas_prices.csv"
df = pd.read_csv(csv_path)
print(df)
```

```python
df['Date'] = pd.to_datetime(df["Date"])
result = df.dtypes
print(result)
```

```python
df2 = df.iloc[530:,:]
df2.head()
```

```python
df2 = df2.reset_index()
df2.head()
```

```python
df2 = df2.drop(labels=["index"], axis=1)
df2.head()
```

```python
df2['Month'] = pd.to_datetime(df2['Date']).dt.to_period('M')
df2 = df2[['Date', 'Month', 'A1', 'A2', 'A3', 'R1', 'R2', 'R3', 'M1', 'M2', 'M3', 'P1', 'P2', 'P3', 'D1']]
df2.head()
```

```python
df2 = df2.drop(labels=['Month','A2', 'A3', 'R1', 'R2', 'R3', 'M1', 'M2', 'M3', 'P1', 'P2', 'P3', 'D1'], axis=1)
df2.head()
```

```python
#df2.groupby(pd.PeriodIndex(df2['Date'], freq="M"))['A1'].mean().reset_index()
df2 = df2.resample('M', on='Date').mean().reset_index()
df2.head()
```

```python
df2['Month'] = pd.to_datetime(df2['Date']).dt.to_period('M')
df2.tail()
```

```python
df2 = df2.drop(labels=['Date'], axis=1)
```

```python
result = df2.dtypes
print(result)
```

```python
df2 = df2.astype({"Month": object})
df2
```

```python
csv_path2 = "wti_cleaned_stock_data.csv"
wti_df = pd.read_csv((csv_path2),index_col=0)
wti_df.head()
```

```python
wti_df = wti_df.iloc[:192,:]
wti_df.tail()
```

```python
result = wti_df.dtypes
print(result)
```

```python
#merge_df = pd.merge(wti_df, df2, on="Month", how="outer")
merge_df = pd.merge(wti_df, df2, left_index=True, right_index=True)
merge_df.head()
```

```python
df3 = merge_df.drop(labels=['Date', 'Month_y'], axis=1)
df3.head()
```

```python
df3 = df3.rename(columns={"Month_x": "Month","A1": "Avg. Gas Price" })
df3.head()
```

```python
df3.to_csv('combined_wti_and_gas.csv')
```
