```python
#Code sheet meant for uploading to Database in PG Admin
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Importing cleaned csv file
gas_cpi_file = "cleaning_results/gas_cpi.csv"
gas_cpi_df = pd.read_csv(gas_cpi_file)
gas_cpi_df.head()
```

```python
#Connecting to local database project2, input in place for password due to public facing code
protocol = 'postgresql'
username = 'postgres'
password = input('<password>')
host = 'localhost'
port = 5432
database_name = 'project2'
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
```

```python
engine.table_names()
```

```python
#Creating name for table in directed database, no index included
gas_cpi_df.to_sql(name='gas_cpi', con=engine, if_exists='replace', index=False)
```

```python
#Test loading of query table from database. Loading succeeded.
pd.read_sql_query('select * from gas_cpi', con=engine).head()
```
