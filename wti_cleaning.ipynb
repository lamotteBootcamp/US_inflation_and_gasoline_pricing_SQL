```python
import pandas as pd
import numpy as np
```

```python
csv_path = "wti_stock_data.csv"
df = pd.read_csv(csv_path)
df.head()
```

```python
df2 = df.drop(df.index[[0, 1, 2, 3]])
```

```python
df2.tail()
```

```python
df3 = df2.drop(labels="Meta Data", axis=1)
df3.head()
```

```python
df3[['Opening Price', 'High Price', 'Low Price', 'Closing Price', 'Volume']] = df3["Monthly Time Series"].apply(lambda x: pd.Series(str(x).split(",")))
df3.head()
```

```python
df4 = df3.drop(labels="Monthly Time Series", axis=1)
df4.head()
```

```python
df4 = df4.rename(columns={"Unnamed: 0": "Date"})
df4 = df4.reset_index()

df4.head()
```

```python
df4 = df4.drop(labels=["index"], axis=1)
df4.head()
```

```python
df4['Date'] = pd.to_datetime(df4["Date"])
print(df4)
```

```python
df4['Opening Price'] = df4['Opening Price'].astype(str)
df4['High Price'] = df4['High Price'].astype(str)
df4['Low Price'] = df4['Low Price'].astype(str)
df4['Closing Price'] = df4['Closing Price'].astype(str)
df4['Volume'] = df4['Volume'].astype(str)
df4.head()
```

```python
result = df4.dtypes
print(result)
```

```python
df4[['Extra', 'Extra1', 'Extra2', 'Opening Price', 'Extra 3']] = df4["Opening Price"].apply(lambda x: pd.Series(str(x).split("'")))

df4.head()
```

```python
df4 = df4.drop(labels=["Extra", "Extra1", "Extra2", "Extra 3"], axis=1)
df4.head()
```

```python
df4[['Extra', 'Extra1', 'Extra2', 'Closing Price', 'Extra 3']] = df4["Closing Price"].apply(lambda x: pd.Series(str(x).split("'")))

df4.head()
```

```python
df4 = df4.drop(labels=["Extra", "Extra1", "Extra2", "Extra 3"], axis=1)
df4.head()
```

```python
df4[['Extra', 'Extra1', 'Extra2', 'High Price', 'Extra 3']] = df4["High Price"].apply(lambda x: pd.Series(str(x).split("'")))

df4.head()
```

```python
df4 = df4.drop(labels=["Extra", "Extra1", "Extra2", "Extra 3"], axis=1)
df4.head()
```

```python
df4[['Extra', 'Extra1', 'Extra2', 'Low Price', 'Extra 3']] = df4["Low Price"].apply(lambda x: pd.Series(str(x).split("'")))

df4.head()
```

```python
df4 = df4.drop(labels=["Extra", "Extra1", "Extra2", "Extra 3"], axis=1)
df4.head()
```

```python
df4[['Extra', 'Extra1', 'Extra2', 'Volume', 'Extra 3']] = df4["Volume"].apply(lambda x: pd.Series(str(x).split("'")))

df4.head()
```

```python
df4 = df4.drop(labels=["Extra", "Extra1", "Extra2", "Extra 3"], axis=1)
df4.head()
```

```python
df4.sort_values(
    by = ['Date'],
    axis = 0, 
    ascending = False, 
    inplace = False, 
    kind = 'quicksort', 
    na_position = 'last', 
    ignore_index = False, 
)

df4.head()
```

```python
result = df4.dtypes
print(result)
```

```python

df4.sort_values(by='Date', inplace=True)
print(df4)
```

```python
df4 = df4.reset_index()
df4 = df4.drop(labels='index', axis=1)
#df4 = df4.drop(labels='level_0', axis=1)
df4.head()
```

```python
df4['Opening Price'] = df4['Opening Price'].astype(float)
df4['High Price'] = df4['High Price'].astype(float)
df4['Low Price'] = df4['Low Price'].astype(float)
df4['Closing Price'] = df4['Closing Price'].astype(float)
df4['Volume'] = df4['Volume'].astype(float)
```

```python
result = df4.dtypes
print(result)
```

```python
df4['Month'] = pd.to_datetime(df4['Date']).dt.to_period('M')
df4 = df4[['Date', 'Month', 'Opening Price', 'High Price', 'Low Price', 'Closing Price', 'Volume']]

df4.head()
```

```python
df4.to_csv('Resources/wti_cleaned_stock_data.csv')
```
